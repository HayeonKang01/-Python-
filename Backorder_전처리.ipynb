# 데이터에 결측치가 너무 많으면?
> ### 결론: 행 순서가 서로에게 영향이 있는 시계열 데이터도 아니고, 결측치의 개수가 굉장히 많기 때문에 그 결측치 값을 예측 모델링으로 예측하여 대체하려 한다.

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("/Users/hayeon/Downloads/team_project_2/dataset/Testing.csv")
df2 = pd.read_csv("/Users/hayeon/Downloads/team_project_2/dataset/Training.csv")
```

```python
# 마지막 행은 데이터가 아니라서 모든 값이 nan이 나오기 때문에 삭제한다. 
df1 = df1.drop(index=242075)
df2 = df2.drop(index=1687860)
```

```python
d = pd.concat([df1, df2])
```

```python
d
```

```python
d.info()
```

```python
for i in range(0,len(d.columns)):
    print(d.columns[i], d[d.columns[i]].unique(), '\n')
```

### lead_time에만 결측치가 존재한다.
### deck_risk, oe_constraint, ppap_risk, stop_auto_buy, rev_stop, went_on_backorder는 Y/N

```python
d[d['lead_time'].isnull() == True]
```

### lead_time : 주문과 배송시간 사이의 간격
> ### lead_time의 결측치의 양이 많아 중간값으로 대체하면 의미가 없어질 듯 하다.  
         때문에 lead_time을 레이블로 잡아 예측으로 결측치를 대체하려 한다.

```python
X_test = d[d['lead_time'].isnull() == True]
```

```python
X_train = d.drop(list(d[d['lead_time'].isnull() == True].index), axis=0)
```

```python
X_test
```

```python
X_test.drop(['lead_time'], axis=1, inplace=True)
```

```python
y_train = X_train[['lead_time']]
X_train.drop(['lead_time'], axis=1, inplace=True)
```

```python
X_train.columns[21]
```

```python
X_train['went_on_backorder'].unique()
```

### No / Yes 로 되어있던 컬럼값들을 No = 0, Yes =  1로 수정

```python
for i in range(16, len(X_train.columns)):
    X_train[X_train.columns[i]].replace({'No':0, 'Yes':1}, inplace=True)
```

```python
for i in range(16, len(X_test.columns)):
    X_test[X_test.columns[i]].replace({'No':0, 'Yes':1}, inplace=True)
```

```python
X_test['potential_issue'].replace({'No':0, 'Yes':1}, inplace=True)
X_train['potential_issue'].replace({'No':0, 'Yes':1}, inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_train) # X_test의 정닶값이 존재하지 않기 때문에(null) X_train으로 정답측정해본다.
```

```python
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_train)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_train)
```

```python
mse_lr = mean_squared_error(y_train, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = r2_score(y_train, y_pred_lr)
```

```python
mse_dt = mean_squared_error(y_train, y_pred_dt)
rmse_dt = np.sqrt(mse_dt)
r2_dt = r2_score(y_train, y_pred_dt)
```

```python
mse_rf = mean_squared_error(y_train, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_train, y_pred_rf)
```

```python
print("< LR > ")
print('mse: ', mse_lr)
print('rmse: ', rmse_lr)
print('r2 : ' , r2_lr)

print("< DT >")
print('mse: ', mse_dt)
print('rmse: ', rmse_dt)
print('r2 : ' , r2_dt)

print("< RF >")
print('mse: ', mse_rf)
print('rmse: ', rmse_rf)
print('r2 : ' , r2_rf)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_train) # X_test의 정닶값이 존재하지 않기 때문에(null) X_train으로 정답측정해본다.

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_train)

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_train)

mse_lr = mean_squared_error(y_train, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = r2_score(y_train, y_pred_lr)

mse_dt = mean_squared_error(y_train, y_pred_dt)
rmse_dt = np.sqrt(mse_dt)
r2_dt = r2_score(y_train, y_pred_dt)

mse_rf = mean_squared_error(y_train, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_train, y_pred_rf)

print("< LR > ")
print('mse: ', mse_lr)
print('rmse: ', rmse_lr)
print('r2 : ' , r2_lr)

print("< DT >")
print('mse: ', mse_dt)
print('rmse: ', rmse_dt)
print('r2 : ' , r2_dt)

print("< RF >")
print('mse: ', mse_rf)
print('rmse: ', rmse_rf)
print('r2 : ' , r2_rf)
```

### DT는 과적합되었다고 판단해, RF가 가장 성능이 좋다고 보인다.
### Test 데이터를 넣어 얻은 결과값으로 결측치를 대체한다.
